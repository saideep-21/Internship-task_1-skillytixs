## Importing the dependent libraries

```python
import pandas as pd
```

importing the dataset from kaggle and converting the .csv into the pandas dataframe

```python
dataset = pd.read_csv('/content/marketing_campaign.csv',header=None)
# printing first 5 rows of dataframe
dataset.head()
```

Splitting the messy data into columns

```python
dataset = dataset[0].str.split('\t',expand=True)
```

finding the shape of the dataset

```python
dataset.shape
```

```python
dataset.head()
```

extracting the column names from dataset present in row 0

```python
columns = dataset.iloc[0,:]
```

Droping the first row from the dataset

```python
dataset.drop(0,axis=0,inplace=True)
```

```python
dataset.head()
```

Changing the columns names

```python
dataset = dataset.rename(columns=columns).reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

Finding the information about each column

```python
dataset.info()
```

Checking for the duplicates in the dataset

```python
dataset[dataset.duplicated()]
```

Checking for the null values in the dataset

```python
dataset.isnull().sum()
```

Converting the columns formats belong to Integers   

```python
string_columns = ['Education','Marital_Status','Dt_Customer']
for column in dataset.columns:
  if column not in string_columns:
    # Convert to numeric, coercing errors (like '') to NaN
    numeric_series = pd.to_numeric(dataset[column], errors='coerce')
    # Fill NaN values (which originated from '') with 0
    filled_series = numeric_series.fillna(0)
    # Convert to integer
    dataset[column] = filled_series.astype(int)
```

```python
dataset.info()
```

```python
dataset['Dt_Customer'] = pd.to_datetime(dataset['Dt_Customer'], format='%d-%m-%Y')
```
